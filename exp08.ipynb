**1.What is 2 tailed and  tailed T Test ?  **

answer->

A t-test is a parametric statistical technique used in Statistics to test hypotheses about population means when:

The sample size is small (n < 30)
The population standard deviation is unknown
The data follows (approximately) a normal distribution

The test is based on the t-distribution, developed by William Sealy Gosset under the pseudonym “Student”. Hence, it is also called Student’s t-test.

A crucial aspect of hypothesis testing using t-tests is deciding whether to use a one-tailed or two-tailed test. This decision is rooted in the nature of the alternative hypothesis.

**1.One-Tailed t-Test (Directional Test):**

A one-tailed t-test is a hypothesis test in which the rejection region lies entirely in one tail of the sampling distribution. It tests for the possibility of the relationship in one specific direction only.

In Hypothesis Testing, every test consists of:


*   Null Hypothesis (H₀): Assumes no effect or no difference
*   Alternative Hypothesis (H₁): Represents the claim we want to test


In a one-tailed test, the alternative hypothesis is directional, meaning:

It predicts not just a difference, but the direction of that difference

*   It predicts not just a difference, but the direction of that difference


 **Types of One-Tailed Tests:**

(A) Right-Tailed Test:


This test evaluates whether the sample mean is significantly greater than the population mean.

Hypotheses Structure:
H₀: μ ≤ μ₀
H₁: μ > μ₀

Interpretation:

If the calculated t-value falls in the right extreme region, it suggests that the observed mean is unusually high under H₀, leading to rejection.

(B) Left-Tailed Test


This test examines whether the sample mean is significantly less than the population mean.

Hypotheses Structure:
H₀: μ ≥ μ₀
H₁: μ < μ₀

Interpretation:

If the t-value lies in the left extreme region, it indicates the sample mean is unusually low.

**Distribution of Significance Level:**

In a one-tailed test:

The entire significance level (α), such as 0.05, is placed in one tail
This increases the critical region size in one direction

**Assumptions:**

1.Observations are independent

2.Data is approximately normally distributed

3.Measurement scale is continuous or interval

4.Random sampling is used

**Applications**

1.Pharmaceutical testing (e.g., drug improves recovery rate)

2.Industrial quality control (e.g., product strength exceeds standard)

3.Educational interventions (e.g., new method increases scores)


**2.Two-Tailed t-Test (Non-Directional Test):**


A two-tailed t-test is a hypothesis test in which the rejection regions are located in both tails of the sampling distribution. It evaluates whether the sample mean is significantly different from the population mean in either direction.

In this test, the alternative hypothesis is non-directional, meaning:

It does not predict whether the difference is positive or negative
It only states that a difference exists:

H₀: μ = μ₀
H₁: μ ≠ μ₀

**Distribution of Significance Level:**

The significance level (α) is divided equally between two tails
For example, if α = 0.05:
0.025 is placed in the left tail
0.025 is placed in the right tail

**Characteristics:**

1.Non-Directional Nature:
The test is neutral and does not assume any specific direction of deviation.

2.Symmetrical Critical Regions
Rejection regions exist on both extremes of the distribution.

3.Lower Power Compared to One-Tailed Test
Because α is split, it becomes harder to detect an effect in one specific direction.

4.Greater Objectivity
It avoids bias and is widely accepted in scientific research.

**Assumptions:**

Same as one-tailed test:

1.Normal distribution

2.Independence of observations

3.Random sampling

4.Interval or ratio scale

**Applications**

1.Scientific experiments where direction is unknown

2.Psychological studies (e.g., treatment has an effect, but unsure of direction)

3.Market research (e.g., change in consumer behavior)

**2.Explain what is Type I and Type II errors with an example.**

answer->

**1.Type I Error (α Error):**


A Type I error occurs when we reject the null hypothesis (H₀) even though it is actually true.

In simple terms:

You conclude that there is an effect or difference, but in reality, there is none.


*   The probability of committing a Type I error is denoted by α (alpha)
*   This is also called the level of significance of the test

*   Common values: α = 0.05 or 0.01


This means:

At α = 0.05, there is a 5% risk of wrongly rejecting a true null hypothesis

**Interpretation:**

Type I error is like a false alarm or false positive:

*  Detecting something that does not actually exist

**Example**

Suppose a pharmaceutical company tests a new drug:

H₀: The drug has no effect

H₁: The drug has an effect

If researchers conclude that the drug works (reject H₀), but in reality it does not, this is a Type I error.


**2.Type II Error (β Error):**

A Type II error occurs when we fail to reject the null hypothesis (H₀) even though it is actually false.

In simple terms:

You conclude that there is no effect, but in reality, there is one.


*   The probability of Type II error is denoted by β (beta)
*   (1 − β) is called the power of the test

*   Power = probability of correctly rejecting a false null hypothesis

**Interpretation:**

Type II error is like a missed detection or false negative:


*   Failing to detect something that truly exists



**Example:**

Using the same drug example:

H₀: The drug has no effect

H₁: The drug has an effect

If researchers conclude that the drug does not work (fail to reject H₀), but in reality it actually works, this is a Type II error.

**3.What is power of statistical test?**

answer->

The power of a statistical test is the probability of correctly rejecting a false null hypothesis (H₀).

 In simple words:

It measures the test’s ability to detect a real effect when it actually exists.

**Mathematical Expression**

Power=1−β


Where:


*   β (beta) = Probability of Type II error
*   Therefore, power is the probability of not making a Type II error

In hypothesis testing:



*   If H₀ is true → we should not reject it
*   If H₀ is false → we should reject it

**Example**

Suppose a researcher tests a new medicine:

H₀: The medicine has no effect

H₁: The medicine has an effect



*  If the medicine truly works and the test correctly detects it, this reflects high power.
* If the test fails to detect the effect (even though it exists), it indicates low power (high β).

The function ttest_rel stands for “t-test for related samples”.

It is used when:

*   The two samples are dependent or paired

*   The same subjects are measured twice (before–after)
*   Or observations are naturally matched (e.g., twins, matched groups)

**Input Characteristics of ttest_rel**

The inputs to ttest_rel define the structure and assumptions of the paired test.

**A) Two Paired Samples (a and b)**

*   The function takes two arrays (or lists) of numerical values
*   Usually written as:





Both samples must be:


*  Of equal length
*   Paired observations (each element in a corresponds to one in b)

**(B) Nature of Data**

Data should be:

1.Continuous (interval or ratio scale)


2.Approximately normally distributed (for differences)

3.Free from extreme outliers


This aligns with assumptions of the t-distribution.


(**C) Optional Parameters**

Useful for multi-dimensional data

2.nan_policy:

Defines how missing values (NaN) are handled:

1.'propagate' → returns NaN

2.'omit' → ignores NaN values

3.'raise' → throws error

**(D) Underlying Theoretical Input**

Internally, the function converts the two samples into:

A difference variable (d = a − b)

**Output Characteristics of ttest_rel**

**(A)test Statistic (t-value):**

**(B) p-value:**


Probability of observing the result assuming H₀ is true

 Interpretation:

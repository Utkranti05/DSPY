Answer->

Non-Graphical EDA performed on the Iris Dataset
Non-graphical EDA means exploring data using statistics and numbers only — no plots or charts.

1.Head / Tail Inspection:

df.head() and df.tail() are used to view the first and last few rows of the dataset.
Helps confirm the data loaded correctly and understand its structure.


2.Shape of Dataset:

df.shape → returns (150, 5) — 150 rows and 5 columns.
Tells us the size of the dataset instantly.


3.Data Types Inspection:

df.dtypes or df.info()
Identifies whether each column is float64, int64, or object (string).
In Iris: 4 features are float64, species is object.


4.Statistical Summary (Descriptive Statistics):

df.describe()
For each numerical feature it gives:

Count — number of non-null entries
Mean — average value
Std — standard deviation (spread of data)
Min / Max — range of values
25%, 50%, 75% — quartiles (Q1, Median, Q3)


This is the most important non-graphical EDA step.


5.Checking Missing Values:

df.isnull().sum()
Counts null/NaN values per column.
Result in Iris → 0 missing values in all columns.


6.Value Counts (Class Distribution):

df['species'].value_counts()
Shows how many rows belong to each class:

setosa → 50
versicolor → 50
virginica → 50


Confirms the dataset is perfectly balanced.


7.Unique Values:

df.nunique() — counts distinct values in each column.
df['species'].unique() — lists the 3 class names.


8.Correlation Matrix:

df.corr()
Computes the Pearson correlation coefficient between all numerical features.
Tells us how strongly two features are linearly related (value between -1 and +1).
In Iris, for example: petal_length and petal_width are highly correlated (~0.96).


9.Groupby / Group Statistics:

df.groupby('species').mean()
Computes the mean of each feature per species.
Very useful to see how species differ numerically without any chart.

Multivariate Analysis Graphs in Air Quality Index Dataset:

1.Heatmap (Correlation Matrix):

Plots the correlation between all numerical features at once (AQI Value, CO, Ozone, NO2, PM2.5).
Color intensity shows the strength of relationship — darker = stronger correlation.
Analyzes multiple variables together in one grid.

2.Pair Plot (Scatter Matrix):

Plots every numerical feature against every other feature in a grid.
The diagonal shows the distribution of each individual variable.
Off-diagonal shows scatter plots between pairs.
Since it shows all features together — it is inherently multivariate.


3.Bar Plot with Hue (Grouped Bar Chart):

A bar chart where color (hue) represents a third variable (e.g., AQI Category).
Example: Top 10 most polluted countries colored by AQI Category.
3 variables involved → Country + AQI Value + AQI Category = Multivariate.


4.Scatter Plot with Hue:

A scatter plot between two numerical variables (e.g., PM2.5 vs AQI Value) colored by AQI Category.
The color adds a third dimension making it multivariate.
Helps see clustering patterns across pollution categories.


5.Box Plot with Hue:

Box plots of AQI values grouped and colored by category or country.
Shows distribution of one variable across multiple groups simultaneously.
3 variables → numerical AQI + category + country = Multivariate.


6.Bubble Chart (if used):

X-axis = one pollutant, Y-axis = another pollutant, bubble size = AQI Value, color = Category.
4 variables in a single chart — strongly multivariate.

4.In Wine quality data set can you do non graphical EDA ? Explain your answer.

Answer->

Yes, Non-Graphical EDA can be performed on the Wine Quality dataset.


Techniques of Non-Graphical EDA:

1.Data Inspection:

The very first step is to look at the raw data. We check the first few and last few records of the dataset. This tells us how the data is structured, what values look like, and whether it loaded correctly. For Wine Quality, we can see rows of decimal values for acidity, sugar, alcohol etc. and an integer quality score.

2.Dimensionality Check:

We check the shape of the dataset — meaning how many rows (observations) and columns (features) it contains. The Wine Quality dataset has 4898 rows and 12 columns. This tells us the scale of the data we are working with.

3.Data Type Analysis:

Each column has a data type. We inspect whether features are integer, float, or string (object). In the Wine Quality dataset, 11 features are of type float64 (decimal numbers) and the quality column is int64 (whole number). This is important because statistical operations like mean and standard deviation only apply to numerical types.

4.Descriptive Statistics:
This is the most important technique in Non-Graphical EDA. For every numerical column we calculate:

Count — total number of valid (non-null) entries
Mean — the average value, representing the central tendency
Standard Deviation — how spread out the values are from the mean
Minimum and Maximum — the range of the data
Quartiles (Q1, Median, Q3) — divide data into four equal parts, helping understand spread and skew

For example, the mean alcohol content is approximately 10.51% and mean quality score is approximately 5.87, telling us most wines are of average quality.

5.Missing Value Analysis:

We check whether any column contains null or NaN values. Missing values can distort statistical analysis and model training. In the Wine Quality dataset, the result is zero missing values across all columns, meaning the data is complete and clean.

6.Duplicate Record Detection:

We check whether the dataset contains identical rows. The Wine Quality dataset contains approximately 937 duplicate rows. This is a critical finding because duplicate records can bias statistical summaries and machine learning models if not handled properly.

7.Class Distribution of Target Variable:

We count how many times each unique value of the target variable quality appears. This tells us whether the dataset is balanced or imbalanced. In the Wine Quality dataset, scores 5 and 6 dominate with very few wines scoring 3 or 9. This means the dataset is imbalanced, which is important to know before building any classification model.

8.Correlation Analysis:

Correlation measures the linear relationship between two numerical variables. The value ranges from -1 to +1 where:

+1 means perfect positive relationship
-1 means perfect negative relationship
0 means no linear relationship


All of this is derived purely from numbers — no graph required.


10.Skewness and Kurtosis:

Skewness measures the asymmetry of a distribution. A positive skew means most values are low with a few very high values. In Wine Quality, residual sugar is highly positively skewed — most wines have low sugar but a few have very high sugar content.
Kurtosis measures the peakedness of a distribution — whether data is heavily concentrated around the mean or spread out.

Both are purely numerical measures requiring no graph.

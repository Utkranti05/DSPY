## 1. Setup & Data Loading

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

**1.Load Dataset**

```python
df = pd.read_csv("movie_reviews_cleaned.csv")

print("Dataset Shape:", df.shape)
print("\nColumns:", df.columns)

df.head()
```

**3.Dataset Information**

```python
df.info()
```

**NON-GRAPHICAL EDA**

(Univariate Analysis)

Information we seek:

Understand the statistical distribution of numerical attributes.

```python
df.describe()
```

**Observation:**

Provides mean, min, max and standard deviation of features.

**Conclusion:**

Helps understand general characteristics of movie reviews.

**Average Review Length by Category (Multivariate)**

Information we seek:

Compare review length between positive and negative reviews.

```python
print(df.columns)
```

```python
df.groupby("category")["word_count"].mean()
```

**Observation**

Shows average word count, character count and sentence count.

**Conclusion**

Helps identify if positive reviews are longer than negative reviews.

GRAPHICAL EDA

**Histogram of Word Count (Univariate)**

Information we seek:

Understand distribution of review length.

```python
plt.figure()

sns.histplot(df["word_count"], bins=30)

plt.title("Distribution of Word Count")
plt.xlabel("Word Count")
plt.ylabel("Frequency")

plt.show()
```

**Observation**

Most reviews fall within a certain word range.

**Conclusion**

Shows the spread of review lengths.

**Box Plot: Word Count vs Category (Multivariate)**

Information we seek:

Compare review length across sentiment categories.

```python
plt.figure()

sns.boxplot(x="category", y="word_count", data=df)

plt.title("Word Count by Review Category")
plt.xlabel("Category")
plt.ylabel("Word Count")

plt.show()
```

**Observation**

Shows median, spread and outliers.

**Conclusion**

Helps compare review sizes between positive and negative reviews.

**1. What is need for Hypothesis Testing ? **

 Need Hypothesis Testing:

 1.To Validate Assumptions:

Before applying any statistical or machine learning model, we make certain assumptions about the data such as normality, equal variance, independence etc. Hypothesis testing helps us verify whether these assumptions are true or false for our dataset.

2.To Make Data-Driven Decisions:

In business, medicine, science and engineering, decisions must be backed by evidence, not intuition. Hypothesis testing provides a formal framework to decide whether an observed effect is real and significant or just a random occurrence.
For example:

Is a new drug actually better than the existing one?
Does a new marketing strategy actually increase sales?


3.To Determine Statistical Significance:

Just because two groups show different averages in a sample does not mean they are truly different in the population. Hypothesis testing tells us whether the difference is statistically significant or simply due to random chance (sampling error).

4.To Compare Groups:

We often need to compare two or more groups and check whether they differ meaningfully.
For example:
Do wines with high alcohol have significantly better quality than wines with low alcohol? (Wine dataset)
Is the AQI of one country significantly higher than another? (AQI dataset)
Do different Iris species have significantly different petal lengths? (Iris dataset)

Without hypothesis testing, such comparisons are just guesswork.

5.To Avoid False Conclusions:

Human observation is prone to bias and error. A pattern visible in a small sample may not exist in reality. Hypothesis testing uses probability and significance levels to protect us from drawing false or misleading conclusions from data.

6.To Measure the Strength of Evidence:

Hypothesis testing gives us a p-value — a numerical measure of how strongly the data supports or contradicts our assumption. This allows us to quantify uncertainty rather than just saying "yes" or "no" based on observation alone.

7.To Support Research and Scientific Inquiry:

In any scientific study or research paper, findings must be proven statistically. Hypothesis testing is the standard accepted method for proving that a result is not due to chance. It gives credibility and reproducibility to research findings.

8.To Identify Relationships Between Variables:

Hypothesis testing helps determine whether a relationship or correlation between two variables is real. For example, in the Wine Quality dataset, is the relationship between alcohol and quality truly significant or just a coincidence in the sample?

3.What is a student t distribution?

answer->

The Student’s t-distribution is a continuous probability distribution that plays a fundamental role in inferential statistics, especially when dealing with small sample sizes. It was developed by William Sealy Gosset, who published his work under the pseudonym “Student,” hence the name Student’s t-distribution.

This distribution is primarily used when the population standard deviation is unknown and the sample size is small (generally less than 30). In such situations, using the normal distribution may not give accurate results because it does not account for the additional uncertainty introduced by estimating the population standard deviation from the sample. The t-distribution adjusts for this uncertainty, making statistical analysis more reliable.

The Student’s t-distribution is bell-shaped, continuous, and symmetric about its mean, which is zero. Although it resembles the normal distribution, it differs in having heavier or thicker tails. These heavier tails indicate that there is a greater probability of obtaining values far from the mean, which reflects the increased variability expected in small samples.

Because of this property, the t-distribution is more spread out than the normal distribution. As a result, it provides more conservative estimates, reducing the risk of underestimating variability in data.

**Applications of Student’s t-Distribution:**

The t-distribution is widely used in various statistical procedures, including:

1.Hypothesis Testing (t-tests)
It is used to test assumptions about population means when sample data is limited. Common types include:


*   Paired sample t-test
*   one-sample t-test
*   Independent sample t-test






2.Confidence Intervals:
It helps in constructing confidence intervals for the population mean when the population standard deviation is unknown.

3.Small Sample Analysis:
It is especially useful in fields like research, engineering, and social sciences where collecting large data samples may not be feasible.

4.What is level of significance ? What role t plays in hypothesis Testing

answer->


**Level of Significance (α):**

The level of significance, denoted by
𝛼
α, is an important concept in statistical hypothesis testing. It is defined as the probability of rejecting the null hypothesis when it is actually true. In other words, it represents the likelihood of committing a Type I error.

In hypothesis testing, we begin by assuming that the null hypothesis  is true. However, based on sample data, we may decide to reject it. Since this decision is based on sample evidence rather than complete population information, there is always a possibility of making an incorrect decision. The level of significance quantifies this risk.

The value of
𝛼
α is chosen by the researcher before conducting the test. Commonly used values are 0.05, 0.01, and 0.10. For example, if the level of significance is set at 0.05, it implies that there is a 5% chance of rejecting the null hypothesis even when it is true. This means that the researcher is willing to accept a 5% risk of making a wrong decision.

The level of significance plays a crucial role in determining the critical region or rejection region of a test. The critical region consists of those values of the test statistic for which the null hypothesis is rejected. The size of this region is directly determined by the value of
𝛼
α. A smaller value of
𝛼
α leads to a narrower critical region, making the test more stringent, whereas a larger value makes the test more lenient.

Thus, the level of significance acts as a threshold for decision-making. It ensures that the conclusions drawn from the test are reliable and that the probability of making an incorrect rejection is controlled. It is widely used in scientific research, quality control, and decision-making processes where uncertainty is involved.

**Role of t in Hypothesis Testing:**

The t-statistic is used in hypothesis testing when the sample size is small and the population standard deviation is unknown. It is based on the Student’s t-distribution, developed by William Sealy Gosset. The t-distribution provides a more appropriate model than the normal distribution in such situations because it accounts for the additional uncertainty involved in estimating the population standard deviation from the sample.

In hypothesis testing, the t-statistic is used to determine whether there is a significant difference between the sample mean and the population mean. It measures how many standard errors the sample mean is away from the hypothesized population mean.


 The degrees of freedom determine the exact shape of the t-distribution and therefore influence the critical value.

If the calculated t-value falls within the critical region, the null hypothesis is rejected, indicating that the sample provides sufficient evidence against it. On the other hand, if the t-value lies outside the critical region, the null hypothesis is not rejected.

The t-statistic plays a central role in various types of hypothesis tests, such as the one-sample t-test, independent sample t-test, and paired sample t-test. These tests are widely used in research to compare means and to draw conclusions about populations based on sample data.

The importance of the t-statistic lies in its ability to provide reliable results even with small samples. By using the t-distribution instead of the normal distribution, it compensates for the additional variability and ensures that the conclusions drawn are statistically valid.

5.Design a hypothesis testing for identifying average age of students in TE IT class is 19 or not.

answer->

Step 1 — Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from scipy import stats
from scipy.stats import t as t_dist

np.random.seed(42)
print("Libraries loaded successfully.")
```

 Step 2 — Sample Data (Ages of TE IT Students

```python
ages = np.array([
    19, 20, 19, 21, 19, 20, 18, 19, 20, 19,
    21, 19, 20, 19, 18, 20, 19, 21, 19, 20
])

print("Student Ages:", ages)
print(f"Number of students (n) : {len(ages)}")
```

 Step 3 — Exploratory Data Analysis (EDA)

```python
n    = len(ages)
xbar = np.mean(ages)
s    = np.std(ages, ddof=1)
se   = s / np.sqrt(n)
df   = n - 1

print("─" * 40)
print(f"  Mean age         (x̄) : {xbar:.4f} years")
print(f"  Std deviation    (s) : {s:.4f}")
print(f"  Standard error  (SE) : {se:.4f}")
print(f"  Sample size      (n) : {n}")
print(f"  Degrees of freedom   : {df}")
print(f"  Min age              : {ages.min()}")
print(f"  Max age              : {ages.max()}")
print(f"  Median age           : {np.median(ages)}")
print("─" * 40)
```

```python
# Visualise the age distribution
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# --- Histogram ---
unique, counts = np.unique(ages, return_counts=True)
axes[0].bar(unique, counts, color='#B5D4F4', edgecolor='#185FA5', width=0.5)
axes[0].axvline(xbar, color='#E24B4A', linewidth=2, linestyle='--',
                label=f'Sample mean = {xbar:.2f}')
axes[0].axvline(19,   color='#1D9E75', linewidth=2, linestyle='-.',
                label='Hypothesised μ₀ = 19')
axes[0].set_xlabel('Age (years)', fontsize=11)
axes[0].set_ylabel('Number of students', fontsize=11)
axes[0].set_title('Age Distribution — TE IT Students', fontsize=12)
axes[0].legend(fontsize=9)
axes[0].grid(axis='y', alpha=0.3)
axes[0].set_xticks(unique)

# --- Box Plot ---
bp = axes[1].boxplot(ages, patch_artist=True, vert=True,
                     medianprops=dict(color='#E24B4A', linewidth=2))
bp['boxes'][0].set_facecolor('#B5D4F4')
bp['boxes'][0].set_edgecolor('#185FA5')
axes[1].axhline(19, color='#1D9E75', linewidth=2, linestyle='-.',
                label='Hypothesised μ₀ = 19')
axes[1].axhline(xbar, color='#E24B4A', linewidth=2, linestyle='--',
                label=f'Sample mean = {xbar:.2f}')
axes[1].set_ylabel('Age (years)', fontsize=11)
axes[1].set_title('Box Plot of Student Ages', fontsize=12)
axes[1].legend(fontsize=9)
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

Step 4 — State Hypotheses & Set Significance Level

```python
mu_0  = 19       # Hypothesised mean (H₀)
alpha = 0.05     # Significance level

print("HYPOTHESIS SETUP")
print("─" * 40)
print(f"  H₀ : μ = {mu_0}  (mean age is 19)")
print(f"  H₁ : μ ≠ {mu_0}  (mean age is NOT 19)")
print(f"  Test type    : Two-tailed (≠)")
print(f"  α            : {alpha}")
print(f"  Conf. level  : {(1-alpha)*100:.0f}%")
print("─" * 40)
```

 Step 5 — Normality Check (Shapiro-Wilk Test)

```python

# Shapiro-Wilk normality test
stat_sw, p_sw = stats.shapiro(ages)

print("NORMALITY CHECK — Shapiro-Wilk Test")
print("─" * 40)
print(f"  W-statistic : {stat_sw:.4f}")
print(f"  p-value     : {p_sw:.4f}")
print("─" * 40)
if p_sw > 0.05:
    print("  Result: p > 0.05 → Data is NORMAL")
    print("  ✓ t-test assumption satisfied")
else:
    print("  Result: p ≤ 0.05 → Data may NOT be normal")
    print("  Consider non-parametric test (Wilcoxon)")
print("─" * 40)
```

```python
# Manual t-statistic calculation
# Formula: t = (x̄ - μ₀) / (s / √n)

t_calc = (xbar - mu_0) / se

# Critical value from t-table at df=n-1, α=0.05, two-tailed
t_crit = t_dist.ppf(1 - alpha / 2, df=df)

# p-value (two-tailed)
p_value = 2 * t_dist.sf(abs(t_calc), df=df)

print("t-TEST COMPUTATION")
print("─" * 45)
print(f"  Formula  : t = (x̄ - μ₀) / (s / √n)")
print(f"           : t = ({xbar:.4f} - {mu_0}) / ({s:.4f} / √{n})")
print(f"           : t = {xbar - mu_0:.4f} / {se:.4f}")
print(f"           : t = {t_calc:.4f}")
print("─" * 45)
print(f"  t-statistic (calculated) : {t_calc:.4f}")
print(f"  t-critical  (±, df={df})  : ±{t_crit:.4f}")
print(f"  p-value     (two-tailed)  : {p_value:.4f}")
print("─" * 45)
```

```python
import numpy as np
from scipy import stats
from scipy.stats import t as t_dist
import matplotlib.pyplot as plt
```

 Sample Data & Statistics

```python
ages  = np.array([19,20,19,21,19,20,18,19,20,19,
                  21,19,20,19,18,20,19,21,19,20])
mu_0  = 19
alpha = 0.05

n    = len(ages)
xbar = np.mean(ages)
s    = np.std(ages, ddof=1)
se   = s / np.sqrt(n)
df   = n - 1

print(f"n={n}  x_bar={xbar:.4f}  s={s:.4f}  SE={se:.4f}  df={df}")
```

 Manual t-Test

```python
# t-statistic
t_calc = (xbar - mu_0) / se

# critical value (two-tailed, alpha=0.05, df=19)
t_crit = t_dist.ppf(1 - alpha/2, df=df)

# p-value
p_val  = 2 * t_dist.sf(abs(t_calc), df=df)

print(f"t_calc  = {t_calc:.4f}")
print(f"t_crit  = +/-{t_crit:.4f}")
print(f"p-value = {p_val:.4f}")
```

3.scipy Verification

```python
t_sp, p_sp = stats.ttest_1samp(ages, popmean=mu_0)

print(f"scipy  t = {t_sp:.4f}")
print(f"scipy  p = {p_sp:.4f}")
print(f"Match with manual: t={np.isclose(t_calc,t_sp)}  p={np.isclose(p_val,p_sp)}")
```

```python
print("-" * 45)
print(f" |t_calc| = {abs(t_calc):.4f}  vs  t_crit = {t_crit:.4f}")
print(f" p-value  = {p_val:.4f}  vs  alpha  = {alpha}")
print("-" * 45)

if p_val <= alpha:
    print(" REJECT H0")
    print(f" Mean age {xbar:.2f} is significantly NOT equal to {mu_0}")
    print(f" at 95% confidence level.")
else:
    print(" ACCEPT H0 (Fail to Reject)")
    print(f" No significant difference from {mu_0}.")
print("-" * 45)
```

```python
ci = stats.t.interval(0.95, df=df, loc=xbar, scale=se)

print(f"95% CI  = ({ci[0]:.4f}, {ci[1]:.4f})")
print(f"mu_0={mu_0} inside CI? {ci[0] <= mu_0 <= ci[1]}")
print("If mu_0 is OUTSIDE CI --> Reject H0  (consistent with p-value)")
```

 6.Plot


```python
x = np.linspace(-5, 5, 500)
y = t_dist.pdf(x, df=df)

fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(x, y, color='#444441', lw=2)

# Rejection regions
ax.fill_between(x, y, where=(x <= -t_crit), color='#E24B4A', alpha=0.4, label='Rejection region')
ax.fill_between(x, y, where=(x >=  t_crit), color='#E24B4A', alpha=0.4)

# Acceptance region
ax.fill_between(x, y, where=((x > -t_crit) & (x < t_crit)),
                color='#B5D4F4', alpha=0.35, label='Acceptance region')

# Lines
ax.axvline( t_crit, color='#1D9E75', ls='--', lw=1.8, label=f't_crit = +/-{t_crit:.3f}')
ax.axvline(-t_crit, color='#1D9E75', ls='--', lw=1.8)
ax.axvline(t_calc,  color='#EF9F27', ls='-',  lw=2.5, label=f't_calc = {t_calc:.3f}')

ax.text(t_calc+0.1, 0.05, f'p={p_val:.4f}', fontsize=9, color='#854F0B')
ax.set_xlabel('t value')
ax.set_ylabel('Density')
ax.set_title(f'One-Sample t-Test | H0: mu=19 | alpha=0.05 | df={df}')
ax.legend(fontsize=9)
ax.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```

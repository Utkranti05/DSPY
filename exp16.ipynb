**1.Problem Statement**

**ML Problem:** Sentiment Classification

**Objective:**
Build a machine learning model that predicts whether a movie review is positive or negative based on its text.

**Type of ML Problem:** Supervised Learning → Classification

**Target Variable:** category (pos / neg)

**2.Import Libraries**

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

**3.Load Dataset**

```python
import nltk
nltk.download('movie_reviews')
```

```python
from nltk.corpus import movie_reviews
import pandas as pd

documents = []

for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        text = movie_reviews.raw(fileid)
        documents.append((fileid, category, text))

df = pd.DataFrame(documents, columns=["file_id", "category", "raw_text"])

df.head()
```

```python
df["word_list"] = df["raw_text"].apply(lambda x: x.split())
df["word_count"] = df["word_list"].apply(len)

df.head()
```

**4.Define Features and Target**

Feature: raw_text

Target: category

```python
X = df["raw_text"]
y = df["category"]
```

**5.Convert Text into Numerical Features**

Machine learning models cannot understand text directly, so we use TF-IDF Vectorization.

```python
vectorizer = TfidfVectorizer(stop_words="english", max_features=5000)

X_vectorized = vectorizer.fit_transform(X)
```

**6.Train-Test Split**

Split dataset into training and testing sets.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_vectorized,
    y,
    test_size=0.2,
    random_state=42
)

print("Training samples:", X_train.shape)
print("Testing samples:", X_test.shape)
```

**7.Train Machine Learning Model**

We use Logistic Regression classifier.

```python
model = LogisticRegression()

model.fit(X_train, y_train)
```

**8.Model Prediction**

```python
y_pred = model.predict(X_test)
```

**9.Model Evaluation**

Accuracy

```python
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)
```

Classification Report

```python
print(classification_report(y_test, y_pred))
```

Confusion Matrix

```python
print(confusion_matrix(y_test, y_pred))
```

Concatenation, or joining of two arrays in NumPy, is primarily accomplished using the routines np.concatenate, np.vstack, and np.hstack. np.concatenate takes a tuple or list of arrays as its first argument, as we can see here:

**Introducing UFuncs**

For many types of operations, NumPy provides a convenient interface into just this kind of statically typed, compiled routine. This is known as a vectorized operation. This can be accomplished by simply performing an operation on the array, which will then be applied to each element. This vectorized approach is designed to push the loop into the compiled layer that underlies NumPy, leading to much faster execution.

```python
np.random.seed(0)

def compute_reciprocals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output

big_array = np.random.randint(1, 100, size=1000000)
%timeit compute_reciprocals(big_array)
```

```python
%timeit (1.0 / big_array)
```

Vectorized operations in NumPy are implemented via ufuncs, whose main purpose is to quickly execute repeated operations on values in NumPy arrays. UFuncs are extremely flexible – before we saw an operation between a scalar and an array, but we can also operate between two arrays as well as multidimentional arrays.

Computations using vectorization through ufuncs are nearly always more efficient than their counterpart implemented using Python loops, especially as the arrays grow in size. Any time you see such a loop in a Python script, you should consider whether it can be replaced with a vectorized expression.

**Exploring NumPy's UFuncs**

Ufuncs exist in two flavors: unary ufuncs, which operate on a single input, and binary ufuncs, which operate on two inputs.

**4.Consider array object named Obj with dimentions 3*4* 2 and you want to convert it to 4*6 shape exlpain how will you do it ? Show this using a code cell, taking an example of 3*4*2 np array of random int between 1 and 50.**

Ans :

To convert an array with dimensions 3x4x2 into a 4x6 shape, you must first ensure the total number of elements remains the same. The original array has \(3\times 4\times 2=24\) elements.The target shape, 4x6, also has \(4\times 6=24\) elements.Because the total element count matches, this reshaping is possible.

```python
import numpy as np

# 1. Create the initial 3*4*2 array with random integers (1-50)
initial_shape = (3, 4, 2)
# Set a seed for reproducibility of the random numbers in this example
np.random.seed(42)
original_array = np.random.randint(1, 51, size=initial_shape)

print(f"Original array shape: {original_array.shape}")
print("Original array (first 2 rows):\n", original_array[:2], "\n")

# 2. Flatten the array to a 1D array
# Both ravel() and flatten() work, ravel() is often more efficient for large arrays
flattened_array = original_array.ravel()
print(f"Flattened array shape: {flattened_array.shape}")
print("Flattened array:\n", flattened_array, "\n")

# 3. Reshape the flattened array to the desired 4*6 shape
target_shape = (4, 6)
reshaped_array = flattened_array.reshape(target_shape)

print(f"Reshaped array shape: {reshaped_array.shape}")
print("Reshaped 4*6 array:\n", reshaped_array)
```

Shallow Copy (e.g., using ... spread syntax): When a nested object (subObjEct) is modified in the shallow copy, the original object is also impacted.
